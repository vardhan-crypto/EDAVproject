```python
# MOVIES DATA ANALYSIS PROJECT — GOOGLE COLAB (improved)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
import io

sns.set(style="whitegrid")  # keep style, avoid passing deprecated palette args

print("Please upload your movies_data.csv file")
uploaded = files.upload()

# === Read file safely ===
file_name = list(uploaded.keys())[0]
df = pd.read_csv(io.BytesIO(uploaded[file_name]))

# Work on an explicit copy to avoid chained-assignment ambiguity
df = df.copy()

print("\nFile loaded successfully!\n")
print("First 5 rows:")
display(df.head())

print("\nDataset info:")
df.info()

# === Basic sanity checks / conversions ===
# Ensure key columns exist
required_cols = {'title', 'rating', 'genre', 'reviews_count', 'release_year'}
missing = required_cols - set(df.columns)
if missing:
    raise ValueError(f"Missing required columns: {missing}")

# Convert numeric-like columns to numeric, coercing bad entries to NaN
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['reviews_count'] = pd.to_numeric(df['reviews_count'], errors='coerce').fillna(0).astype(int)
df['release_year'] = pd.to_numeric(df['release_year'], errors='coerce').astype('Int64')  # nullable integer

# STEP 2: DATA CLEANING (avoid inplace on slices)
# Handle missing ratings by replacing NaN with mean rating (computed from valid ratings)
mean_rating = df['rating'].mean(skipna=True)
df['rating'] = df['rating'].fillna(mean_rating)

# Replace blank or whitespace-only genres with NaN then fill
df['genre'] = df['genre'].astype(str).str.strip()
df.loc[df['genre'] == '', 'genre'] = np.nan
df['genre'] = df['genre'].fillna('Unknown')

# === Q1: Average rating by genre ===
avg_rating_by_genre = df.groupby('genre', sort=False)['rating'].mean().sort_values(ascending=False)
print("\nQ1: Average Rating by Genre:")
print(avg_rating_by_genre)

# === Q2: Most reviewed movies (top 10) ===
most_reviewed = df.sort_values(by='reviews_count', ascending=False).head(10)
print("\nQ2: Top 10 Most Reviewed Movies:")
display(most_reviewed[['title', 'reviews_count', 'rating']])

# === Q3: Verify Unknown genres count ===
unknown_count = int((df['genre'] == 'Unknown').sum())
print(f"\nQ3: Number of movies with genre='Unknown': {unknown_count}")

# === Q4: Rating trends across decades ===
# Create decade column safely; handle missing release_year
df['decade'] = df['release_year'].dropna().astype(int) // 10 * 10
# For rows where release_year is missing, put NaN in decade
df.loc[df['release_year'].isna(), 'decade'] = pd.NA

decade_trends = df.dropna(subset=['decade']).groupby('decade')['rating'].mean().sort_index()
print("\nQ4: Average Rating by Decade:")
print(decade_trends)

# === Q5: Visualizations ===
# 1) Genre distribution — use pandas plotting to avoid seaborn palette/hue warnings
plt.figure(figsize=(14, 6))
genre_counts = df['genre'].value_counts()
genre_counts.plot(kind='barh')
plt.title("Genre Distribution")
plt.xlabel("Count of Movies")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()

# 2) Rating comparison by genre — use seaborn boxplot but avoid passing 'palette' without hue
plt.figure(figsize=(12, 8))
# Order genres by median rating for better readability
genre_order = df.groupby('genre')['rating'].median().sort_values(ascending=False).index
sns.boxplot(x='rating', y='genre', data=df, order=genre_order)
plt.title("Rating Comparison by Genre")
plt.xlabel("Rating")
plt.ylabel("Genre")
plt.tight_layout()
plt.show()

# 3) Rating trends across decades
plt.figure(figsize=(10, 5))
decade_trends_plot = decade_trends.reset_index().dropna()
sns.lineplot(x='decade', y='rating', data=decade_trends_plot, marker='o')
plt.title("Average Rating Trends Across Decades")
plt.xlabel("Decade")
plt.ylabel("Average Rating")
plt.grid(True)
plt.tight_layout()
plt.show()

print("\nData Analysis Completed Successfully! (Q1–Q5 answered with cleaned code and visualizations.)")
```
